```python
# 1. set a random seed for reproducilbility
np.random.seed(42)

# 2. generate 100 random float values between 10 to 100
data = np.random.uniform(10,100, size =100)

print(data)
```

```python
# 3. Randomly insert 10 NaN values
nan_indices = np.random.choice(data.size, size=10, replace=False)
data[nan_indices] = np.nan
```

```python
# 4. Randomly insert 5 outliers( e.g., values>200)
outlier_indices = np.random.choice(np.setdiff1d(np.arange(data.size), nan_indices), size=5, replace=False)
data[outlier_indices] = np.random.uniform(200,300, size = 5)
```

```python
# 5. Pritn the first 15 values to insepct
print(data[:15])
```

```python
print(data)
```

```python
#Lets Clean the Data

#Step 1 : remove nan values
cleaned_data = data[~np.isnan(data)]

#Step 2: remvoe outliers(number greater than 150)
cleaned_data = cleaned_data[cleaned_data<150]

#Step 3: compute basic statistics
mean_val = np.mean(cleaned_data)
median_val = np.median(cleaned_data)
std_dev = np.std(cleaned_data)

#Step 4: Print the results
print(F"Cleaned Data Statistics:")
print(f"Mean :{mean_val:.2f}")
print(f"Median :{median_val:.2f}")
print(f"Standard Deviation :{std_dev:.2f}")
print(cleaned_data)
```

```python
import matplotlib.pyplot as plt

# plot the cleaned Data

plt.figure(figsize = (10,5))
plt.plot(cleaned_data, marker='o', linestyle='-', color='blue')
plt.title("Cleaned Dataset - Trend Visualization")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
