```python
# 1. set a random seed for reproducilbility
np.random.seed(42)

# 2. generate 100 random float values between 10 to 100
data = np.random.uniform(10,100, size =100)

print(data)
```

```python
# 3. Randomly insert 10 NaN values
nan_indices = np.random.choice(data.size, size=10, replace=False)
data[nan_indices] = np.nan
```

```python
# 4. Randomly insert 5 outliers( e.g., values>200)
outlier_indices = np.random.choice(np.setdiff1d(np.arange(data.size), nan_indices), size=5, replace=False)
data[outlier_indices] = np.random.uniform(200,300, size = 5)
```

```python
# 5. Pritn the first 15 values to insepct
print(data[:15])
```
